# CALIFORNIA

```python
# Filter columns based on Geography - California - Standard - Multi Outlet + Conv
filtered_df1 = df1[df1['Geography'] == 'California - Standard - Multi Outlet + Conv']
filtered_df2 = df2[df2['Geography'] == 'California - Standard - Multi Outlet + Conv']
filtered_df3 = df3[df3['Geography'] == 'California - Standard - Multi Outlet + Conv']
filtered_df4 = df4[df4['Geography'] == 'California - Standard - Multi Outlet + Conv']
filtered_df5 = df5[df5['Geography'] == 'California - Standard - Multi Outlet + Conv']
```

```python
df1['Geography'].value_counts()
```

```python
# Aggregate the sales for each name by summing up
sales_by_name = filtered_df1.groupby('Brand Franchise Name')['Unit Sales'].sum().reset_index()
# This will give you a DataFrame where each 'Name' is unique and 'Sales' represents the total sales for that name
```

```python
# Filter Brand with total sales greater than 200,000
filtered_sales_by_name = sales_by_name[sales_by_name['Unit Sales'] > 200000]
# Use the filtered aggregated data for plotting
names = filtered_sales_by_name['Brand Franchise Name']
total_sales = filtered_sales_by_name['Unit Sales']

plt.figure(figsize=(15, 8))  # Adjust figure size for visibility

# Iterate over names and sales, highlighting 'X'
for name, sale in zip(names, total_sales):
    if name == 'GARDEIN':  # Check if the name is 'X'
        plt.bar(name, sale, color='red')  # Highlight 'X' with a different color, e.g., red
    else:
        plt.bar(name, sale, color='skyblue')  # Use a default color for other bars

plt.title('Total Sales by Brand (Sales > 200,000)')
plt.xlabel('Brand Franchise Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)  # Rotate names for readability
plt.grid(axis='y')

plt.show()
```

```python
#combining 2023 and 2024 data
combined_df = pd.concat([filtered_df4, filtered_df5], ignore_index=True)
```

```python
#combined sales data for 2023 and 2024
sales_by_name_combined = combined_df.groupby('Brand Franchise Name')['Unit Sales'].sum().reset_index()
#filter data -> sales value >200,000
filtered_sales_by_name_combined = sales_by_name_combined[sales_by_name_combined['Unit Sales'] > 200000]
```

```python
# Plot the graph

names = filtered_sales_by_name_combined['Brand Franchise Name']
total_sales = filtered_sales_by_name_combined['Unit Sales']

plt.figure(figsize=(15, 8))
for name, sale in zip(names, total_sales):
    if name == 'GARDEIN':  # Highlight 'X' with a different color
        plt.bar(name, sale, color='red')
    else:
        plt.bar(name, sale, color='skyblue')

plt.title('Total Sales (2023+2024) (Sales > 200,000)')
plt.xlabel('Brand Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.grid(axis='y')

plt.show()
```
